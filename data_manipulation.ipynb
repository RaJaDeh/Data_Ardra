### Standardize Words

```python
fat_map = {

    'lf' : 'Low Fat',
    'reg' : 'Regular',
    'low fat' : 'Low Fat',
    'regular' : 'Regular'

}
```

```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].map(lambda x: fat_map.get(x.lower(), x))
```

### Item Wieght Null Handling

```python
df
```

```python
df.Item_Weight.describe()
```

```python
df.sort_values('Item_Identifier')
```

```python
# Buat mapping dari Item_Identifier ke satu nilai Item_Weight yang diketahui (bukan NaN)
weight_map = df.loc[df['Item_Weight'].notna()].drop_duplicates('Item_Identifier').set_index('Item_Identifier')['Item_Weight']

# Isi NaN berdasarkan mapping ini
df['Item_Weight'] = df['Item_Weight'].fillna(df['Item_Identifier'].map(weight_map))
```

```python
df.dropna(inplace=True)
```
